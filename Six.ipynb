```python
sample_submission = pd.read_csv(r"C:\Users\india\Desktop\Projects\Flask_App\House Price Prediction\sample_submission.csv")
test = pd.read_csv(r"C:\Users\india\Desktop\Projects\Flask_App\House Price Prediction\test.csv")
```

```python
dataset1 = pd.merge(test, sample_submission, how ='inner', on ='Id')
```

```python
features_with_na=[features for features in dataset1.columns if dataset1[features].isnull().sum()>1 and dataset1[features].dtypes=='O']

for feature in features_with_na:
    print(feature, np.round(dataset1[feature].isnull().mean()*100, 4),  ' % missing values')
```

```python
def replace_cat_feature(dataset1,features_with_na):
    data=dataset1.copy()
    data[features_with_na]=data[features_with_na].fillna('Missing')
    return data

dataset1=replace_cat_feature(dataset1,features_with_na)

dataset1[features_with_na].isnull().sum()
```

```python
numerical_with_nan=[feature for feature in dataset1.columns if dataset1[feature].isnull().sum()>1 and dataset1[feature].dtypes!='O']
numerical_with_nan
for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(dataset1[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=dataset1[feature].median()
    
    ## create a new feature to capture nan values
    dataset1[feature+'nan']=np.where(dataset1[feature].isnull(),1,0)
    dataset1[feature].fillna(median_value,inplace=True)
    
dataset1[numerical_with_nan].isnull().sum()
    
    
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    dataset1[feature]=dataset1['YrSold']-dataset1[feature]
```

```python
dataset1.head()
```

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset1[feature]=np.log(dataset1[feature])
```

```python
dataset1.head(100)
```

```python
categorical_features=[feature for feature in dataset1.columns if dataset1[feature].dtype=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset1.groupby(feature)['SalePrice'].count()/len(dataset1)
    temp_df=temp[temp>0.01].index
    dataset1[feature]=np.where(dataset1[feature].isin(temp_df),dataset1[feature],'Rare_var')  
```

```python
dataset1.head(100)
```

```python
for feature in categorical_features:
    labels_ordered=dataset1.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset1[feature]=dataset1[feature].map(labels_ordered)
```
